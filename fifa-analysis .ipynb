### Renaming columns with appropriate format

```python
df.columns
```

```python
df = df.rename(columns = {"short_name":"Short_Name",
                          "player_positions":"Position",
                          "value_eur":"Value_Eur",
                          "wage_eur":"Wage_Eur",
                          "age":"Age",
                          "height_cm":"Height_cm",
                          "weight_kg":"Weight_kg",
                          "club_name":"Club_Name",
                          "league_name":"League_Name",
                          "nationality_name":"Nationality",
                          "preferred_foot":"Preferred_Foot",
                          "pace":"Pace",
                          "shooting":"Shooting",
                          "passing":"Passing",
                          "dribbling":"Dribbling",
                          "defending":"Defending",
                          "physic":"Physic",
                          "overall":"Overall"
                         })
```

```python
df.columns
```

### Creating new feature is_loaned from the help of column club_loaned_from.

```python
df.head()
```

```python
df["Is_Loaned"] = df["club_loaned_from"].map(lambda x:0 if pd.isna(x) else 1)
```

#### Dropping the old column

```python
df.drop(columns=["club_loaned_from"],inplace=True)
```

### creating new feeature No_Position representing number of position any player could play

```python
df["No_Position"] = df["Position"].str.split(",").str.len()
```

```python
df["No_Position"].head()
```

### Handling missing values

```python
df.isna().sum()
```

#### There is common 61 rows with missing values of multiple features. Can be dropped as well but filling those missing values is possible.

```python
df.loc[df["Wage_Eur"].isna()]
```

```python
df["Club_Name"] = df["Club_Name"].fillna("Free Agent")
df["League_Name"] = df["League_Name"].fillna("No League")

df["Value_Eur"] = df.groupby(["Overall","Position"])["Value_Eur"].transform(lambda x:x.fillna(x.median()))
df["Wage_Eur"] = df.groupby(["Overall","Position"])["Wage_Eur"].transform(lambda x:x.fillna(x.median()))
```

```python
df.isna().sum()
```

#### There is still very small missing value left for those columns. This might due to less data of grouping of combining two features. so only grouping with 1 feature for remaining missing values.

```python
df["Value_Eur"] = df.groupby("Overall")["Value_Eur"].transform(lambda x:x.fillna(x.median()))
df["Wage_Eur"] = df.groupby("Overall")["Wage_Eur"].transform(lambda x:x.fillna(x.median()))
```

```python
df.isna().sum()
```

### There is still missing values of 2132 rows for multiple features. Figuring out those features.

```python
df.loc[df.Pace.isna()]
```

```python
df.loc[(df.Pace.isna()) & (df["Position"]=="GK")].shape
```

### Figured out that all the 2132 missing values are of position GK

#### Splitting the dataset into two. For GK and for Players other than GK. Splitting is carried out inorder to analyze the GK analysis seperately than other players analysis

```python
df_gk = df[df["Position"].str.contains("GK")]
df_outfield = df[~df["Position"].str.contains("GK")]
```
